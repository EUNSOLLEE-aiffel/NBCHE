```python
# Step 1. 파일 불러오기
file = open("Avengers.txt")  #MyDrive에서 Avengers.txt를 불러오기
contents = file.read()
print(contents)
```

```python
# Step 2. 내용 정리하기
## Step 2.1 줄바꿈제거 / 소문자로 변경
revised_1 = contents.strip("\n").lower()

## Step 2.2 특수문자 제거
import string
revised_2 = revised_1.translate(str.maketrans('', '', string.punctuation))
print(revised_2)
```

```python
# Step 3. 단어 정리하기
## Step 3.1 단어 마다 리스트작업
text_1 = revised_2.split()
print(text_1)
print(len(text_1))
text_2 = list(set(text_1))
print(len(text_2))

## Step 3.2 단어별 빈도수를 포함한 딕셔너리 생성
text_3 = {text_2[i]:text_1.count(text_2[i]) for i in range(len(text_2))}
print(text_3)
print(len(text_3))

## Step 3.3 빈도수로 정렬
text_4_d = dict(sorted(text_3.items(), key=lambda x: x[1], reverse=True))
print(text_4_d)
print(len(text_4_d))
```

```python
# Step 4. 순위 인덱스 생성
text_5_d = {}
for i, (k, v) in enumerate(text_4_d.items()):
    text_5_d[k] = (i + 1, v) # Assign the rank (i+1) and frequency (v) to the word (k) in the dictionary

print(text_5_d)
```

```python
# Step 5. 입력 단어에 맞게 결과 보여주기
input_text = list(input('단어 또는 문장을 입력하세요:').split())
a = [x for x in input_text]
print(a)

b = [text_5_d.get(x)[0] for x in a]
print(b)
```
